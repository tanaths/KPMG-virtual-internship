# **Customer Demographic**

```python
customer_demography = pd.read_excel('/content/drive/MyDrive/Internship/KPMG_VI_New_raw_data_update_final.xlsx', sheet_name='CustomerDemographic')
customer_demography
```

```python
customer_demography.describe(include='all')
```

The gender column has inconsistent format such as 'F', 'Femal', 'M', and 'U'. Now we can change it into proper format

```python
customer_demography['gender'].replace(['Femal', 'F'], 'Female', inplace=True)
customer_demography['gender'].replace('M', 'Male', inplace=True)
customer_demography['gender'].replace('U', 'Unknown', inplace=True)
```

Based on describe column, we see that there is inaccurate format of DOB (Day of Birth), it is 1843-12-21. which counts as inacurrate data, probably an human error. Let's change it into 1943

```python
customer_demography['DOB'] = customer_demography['DOB'].replace('1843-12-21', '1943-12-21')
```

drop default column as it has irrelevant data

```python
customer_demography = customer_demography.drop('default', axis = 1)
```

let's count number of missing values

```python
customer_demography.isnull().sum()
```

> we will handle missing value with dropping and imputation. for last_name we wont do anything since it wont have significant impact. But for job_title, job_industry_category we will have imputation on it. NaN on Tenure and DOB will be dropped

```python
customer_demography.duplicated().sum()
```

```python
#fill the missing values
customer_demography['job_title'].fillna(value='Unknown', inplace=True)
customer_demography['job_industry_category'].fillna(value='Unknown', inplace=True)
```

```python
#drop missing values
customer_demography.dropna(subset=['DOB', 'tenure'], inplace=True)
```

```python
customer_demography.isnull().sum()
```

**For age segmentation, we will add new columns based on DOB**

```python
from datetime import datetime

def calculate_age(dob):
    current_date = datetime.now()
    age = current_date.year - dob.year - ((current_date.month, current_date.day) < (dob.month, dob.day))
    return age

customer_demography['age'] = customer_demography['DOB'].apply(calculate_age)
```

grouping age by age category

```python
bins = [0,20,30,40,50,60,100]
labels = ['10-20', '21-30', '31-40','41-50','51-60','61+']

customer_demography['age_category'] = pd.cut(customer_demography['age'], bins=bins, labels=labels, right=False)
grouped_age = customer_demography.groupby('age_category').size().reset_index(name='count')
```

As we can see, product_first_sold_date has wrong data type. We will change it into [datetime]

```python
transactions.info()
```

Let's checking missing value

```python
transactions.isnull().sum()
```

since it has so many missing values, we will drop it. It will be fitered out

```python
transactions.dropna(subset=['online_order', 'brand','product_line', 'product_class',
                            'product_size', 'product_line', 'standard_cost',
                            'product_first_sold_date'], inplace=True)
```

```python
transactions.isnull().sum()
```

```python
transactions
```

```python
transactions['product_first_sold_date'].head(20)
```

extracts only the date part from the datetime values

```python
transactions['product_first_sold_date'] = transactions['product_first_sold_date'].dt.date
```

```python
transactions['product_first_sold_date'].head(20)
```

The date part is correct but unfortunately the data type changed to object, so let's change it again into datetime format

```python
transactions['product_first_sold_date'] = pd.to_datetime(transactions['product_first_sold_date'], errors='coerce')
```

```python
transactions.info()
```

**Let's calculate profit based on standard cost and price**

```python
transactions['profit'] = transactions['list_price'] - transactions['standard_cost']
transactions['profit'] = transactions['profit'].apply(lambda x: round(x,2))
transactions
```
